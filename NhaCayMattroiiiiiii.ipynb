```python
import turtle

t = turtle.Turtle()
turtle.bgcolor("Black")

t.speed(0)

t.pensize(5)

#Vẽ nhà
t.penup()
t.lt(180)
t.fd(300)
t.pendown()
t.lt(180)

#Mái nhà Đỏ
t.begin_fill()
t.color("white","red")
for i in range(3):
    t.fd(150)
    t.lt(120)
t.end_fill()

#Thân hồng
t.begin_fill()
t.color("white","pink")
for i in range(4):
    t.fd(150)
    t.rt(90)
t.end_fill()

#Mái xanh
t.begin_fill()
t.color("white","blue")
t.fd(150)
t.lt(55)
t.fd(100)
t.lt(65)
t.fd(150)
t.lt(115)
t.fd(100)
t.lt(65)
t.fd(150)
t.end_fill()

#Thân xám
t.begin_fill()
t.color("white","gray")
t.rt(30)
t.fd(150)
for i in range(2):
    t.lt(145)
    t.fd(100)
    t.lt(35)
    t.fd(150)
t.lt(90)
t.end_fill()

#Mặt trời
t.speed(0)
t.penup()
t.fd(350)
t.lt(90)
t.fd(330)
t.rt(90)
t.pendown()
t.pensize(3)
turtle.bgcolor("black")
t.hideturtle()

color = ["red","cyan","yellow","pink","orange","green","white","darkorchid"]
for j in range(75):
    t.pencolor(color[j%8])
    t.width(j/100+1)
    t.fd(j)
    t.lt(59)

#Cây

from turtle import *
from random import *
turtle.pensize(3)
#turtle.penup()

#turtle.pendown()
rt(90)
fd(200)
lt(90)
fd(150)
lt(90)
pencolor("chocolate")

def branch(branch_len):
  angle = randint(25,27)
  pencolor("chocolate")
  if branch_len > 10:
    fd(branch_len)
    lt(angle)
    branch(branch_len*0.7)
    rt(angle*2)
    branch(branch_len*0.7)
    lt(angle)
    backward(branch_len)

branch(100)
```
